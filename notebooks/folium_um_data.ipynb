```python
event_coordinates = folium_um_data.groupby('Event name').first().reset_index()
```

```python
average_lat = event_coordinates['latitude'].mean() 
average_lon = event_coordinates['longitude'].mean() 
```

### Add Custom Markers

```python
# Create a map 
m = folium.Map(location=[39.50, -98.35], zoom_start=4) 
# Add custom markers 
for lat, lon, name in zip(event_coordinates['latitude'], event_coordinates['longitude'], event_coordinates['Event name']): 
    folium.Marker( location=[lat, lon], popup=name, icon=folium.Icon(color='red', icon='info-sign') ).add_to(m)
```

```python
mult_map = folium.Map(location=[39.50, -98.35], zoom_control=False, control_scale=True, zoom_start=4)

colormap = plt.cm.get_cmap('YlOrRd')
normalize = plt.Normalize(vmin=event_coordinates['Distance (miles)'].min(), vmax=event_coordinates['Distance (miles)'].max())


for i in range(len(event_coordinates['latitude'])):
    distance = event_coordinates['Distance (miles)'][i]
    color= colormap(normalize(distance))
    color_hex= mcolors.rgb2hex(color[:3])

    popup_text = f"Name: {event_coordinates['Event name'][i]}<br>Elevation Gain: {event_coordinates['Elevation Gain (feet)'][i]}<br>Distance: {distance}"
    folium.Marker(
        location=[event_coordinates['latitude'][i], event_coordinates['longitude'][i]],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(icon_color=color_hex)
    ).add_to(mult_map)
    
unique_distances = event_coordinates['Distance (miles)'].unique() 
legend_html = '<div style="position: fixed; bottom: 50px; left: 50px; width: 200px; height: auto; background-color: white; border:2px solid grey; z-index:9999; font-size:14px; opacity: 0.9; padding: 10px;"><h4>Distance (miles)</h4>' 
for distance in unique_distances: 
    color = colormap(normalize(distance)) 
    color_hex = mcolors.to_hex(color[:3]) 
    legend_html += f'<i style="background-color: {color_hex}; color: white; border-radius: 50%; padding: 5px;">&nbsp;</i> {distance} miles<br>' 
legend_html += '</div>'
mult_map.get_root().html.add_child(folium.Element(legend_html))
```
