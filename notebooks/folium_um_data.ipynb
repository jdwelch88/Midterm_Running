```python
folium_um_data['latitude'] = folium_um_data['Event name'].map(latitude)
folium_um_data['longitude'] = folium_um_data['Event name'].map(longitude)
```

```python
event_coordinates = folium_um_data.groupby('Event name').first().reset_index()
```

```python
average_lat = event_coordinates['latitude'].mean() 
average_lon = event_coordinates['longitude'].mean() 
```

```python


event_map = folium.Map(location=[average_lat, average_lon], control_scale=True, zoom_control=False, zoom_start=4)

# FastMarkerCluster(data=list(zip(event_coordinates['latitude'], event_coordinates['longitude']))).add_to(event_map)
for _, row in event_coordinates.iterrows(): 
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['Event name'], 
    ).add_to(event_map)
title_html = '''
    <h3 align="center" style="font-size:20px"><b>Top 100 Ultra Marathon Events</b></h3>
    '''
event_map.get_root().html.add_child(folium.Element(title_html))
event_map
```
